```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f_sphere(x,y):
    return x**2 + y**2

#define a meshgrid for ploting the function.

x_range = np.arange(-15, 15, 0.1)
y_range = np.arange(-15, 15, 0.1)

X, Y = np.meshgrid(x_range, y_range)
Z = f_sphere(X,Y)
```

```python
#Plot the function
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,Z, cmap='coolwarm', alpha=0.5)
ax.scatter(0,0,0, c='r', marker='o')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('function')
plt.show()
```

# Code done for homework, function 1.

Here we have the libraries used for this homework, one is for the math analysis and the second one is for the plots and graphics used in the code below.

```python
import numpy as np
import matplotlib.pyplot as plt
```

f_sphere is the function used based on the class work that we had, in this home work was defined a quadratic function but for the homework a cubic function was used.

Gradient Function: gradient_f computes the gradient of the function
$$
 \bullet{(,)=^{3}+^{3}}
$$
This gradient is used to update $x$ and $$ in the gradient descent algorithm.

Gradient Descent Algorithm: gradient_descent function performs the optimization by iteratively moving in the direction opposite to the gradient.

Path and Final Point: The path of the descent (all intermediate points) is plotted along with the final optimized point.

Plotting: The path taken during gradient descent is plotted in green, and the final point is marked with a red star.

```python
#Here is were the function that is going to return the x and y parameters is declared, in this example basically we used just the same function as before (for the class work) just with a different exponent.
def f_sphere_4HW(x,y):
    return x**3 + y**3

# Gradient of the function
def gradient_f(x, y):
    dfdx = 3 * x**2
    dfdy = 3 * y**2
    return np.array([dfdx, dfdy])

# Gradient Descent Algorithm
def gradient_descent(starting_point, learning_rate, num_iterations):
    x, y = starting_point
    path = [(x, y)]
    
    for _ in range(num_iterations):
        grad = gradient_f(x, y)
        x -= learning_rate * grad[0]
        y -= learning_rate * grad[1]
        path.append((x, y))
    
    return np.array(path), (x, y)

# Parameters for Gradient Descent
starting_point = (5, 5)  # Initial guess
learning_rate = 0.001
num_iterations = 100000

# Perform Gradient Descent
path, min_point = gradient_descent(starting_point, learning_rate, num_iterations)

#define a meshgrid for ploting the function.

x_range = np.arange(-15, 15, 0.1)
y_range = np.arange(-15, 15, 0.1)

X, Y = np.meshgrid(x_range, y_range)
Z = f_sphere_4HW(X,Y) #Here we call out function before declared and we send the x and y values generated, this values are going to be used to create the graphyc.
```

```python
#Plot the function
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,Z, cmap='coolwarm', alpha=0.5)  # Surface
ax.plot(path[:, 0], path[:, 1], f_sphere_4HW(path[:, 0], path[:, 1]), c='g', marker='o')  # Path of descent
ax.scatter(0,0,0, c='r', marker='*')  # Final optimized point
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Function')
plt.show()
#With all this instructions we basically just generate the figure with all the things we need. 
```

Plotting: The path taken during gradient descent is plotted in green, and the final point is marked with a red star.

```python
#Here is were the function that is going to return the x and y parameters is declared, in this example basically we used just the same function as before (for the class work) just with a different exponent.
def f_sphere_4HW(x,y):
    return np.sin(x) + np.cos(y)

# Gradient of the function
def gradient_f(x, y):
    dfdx = np.cos(x)
    dfdy = -np.sin(y)
    return np.array([dfdx, dfdy])

# Gradient Descent Algorithm
def gradient_descent(starting_point, learning_rate, num_iterations):
    x, y = starting_point
    path = [(x, y)]
    
    for _ in range(num_iterations):
        grad = gradient_f(x, y)
        x -= learning_rate * grad[0]
        y -= learning_rate * grad[1]
        path.append((x, y))
    
    return np.array(path), (x, y)

# Parameters for Gradient Descent
starting_point = (5, 5)  # Initial guess
learning_rate = 0.001
num_iterations = 100000

# Perform Gradient Descent
path, min_point = gradient_descent(starting_point, learning_rate, num_iterations)

#define a meshgrid for ploting the function.

x_range = np.arange(-15, 15, 0.1)
y_range = np.arange(-15, 15, 0.1)

X, Y = np.meshgrid(x_range, y_range)
Z = f_sphere_4HW(X,Y) #Here we call out function before declared and we send the x and y values generated, this values are going to be used to create the graphyc.
```

```python
#Plot the function
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,Z, cmap='coolwarm', alpha=0.3)  # Surface
ax.plot(path[:, 0], path[:, 1], f_sphere_4HW(path[:, 0], path[:, 1]), c='b', marker='*')  # Path of descent
ax.scatter(0,0,0, c='r', marker='*')  # Final optimized point
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Function')
plt.show()
#With all this instructions we basically just generate the figure with all the things we need. 
```
